### Po co wizualizujemy dane?

üïµÔ∏è‚Äç‚ôÇÔ∏è **Odkrywanie wzorc√≥w i relacji**: To g≈Ç√≥wny cel. Szukamy zale≈ºno≈õci miƒôdzy r√≥≈ºnymi zmiennymi (np. czy klienci, kt√≥rzy wiƒôcej kupujƒÖ, mieszkajƒÖ w konkretnym mie≈õcie?). PomagajƒÖ w tym wizualizacje, jak wykresy korelacji.

üßπ **Wykrywanie b≈Çƒôd√≥w i problem√≥w**: Szukamy "brud√≥w" w danych, takich jak:

- Braki danych (puste kom√≥rki).

- Anomalie (warto≈õci odstajƒÖce, np. wiek klienta 150 lat).

- B≈Çƒôdy formatowania (np. "Krak√≥w" i "krakow" jako dwie r√≥≈ºne kategorie).

- Zmienno≈õƒá danych w czasie (np. zmiana kodowania statusu klienta, zapisywanie kwot w z≈Çot√≥wkach / groszach).

üìä **Zrozumienie struktury danych**: Sprawdzamy, jakie mamy kolumny (cechy), jakie sƒÖ typy tych danych (liczby, tekst, daty) i jaki jest rozk≈Çad warto≈õci (np. ilu jest mƒô≈ºczyzn, a ile kobiet).

ü§î **Sprawdzanie za≈Ço≈ºe≈Ñ**: Wiele modeli statystycznych i AI (np. regresja liniowa) dzia≈Ça poprawnie tylko przy pewnych za≈Ço≈ºeniach (np. gdy dane majƒÖ rozk≈Çad normalny). EDA pozwala to zweryfikowaƒá.

üí° **Generowanie hipotez**: Na podstawie obserwacji z EDA mo≈ºemy postawiƒá pierwsze hipotezy, kt√≥re p√≥≈∫niej bƒôdziemy testowaƒá (np. "Wydaje siƒô, ≈ºe promocje weekendowe generujƒÖ wiƒôkszy zysk ni≈º te w tygodniu").

üõ†Ô∏è **Wsparcie dla "Feature Engineering"**: EDA daje nam pomys≈Çy, jak ulepszyƒá dane dla modelu. Mo≈ºemy zdecydowaƒá, kt√≥re kolumny sƒÖ bezu≈ºyteczne, a kt√≥re warto ze sobƒÖ po≈ÇƒÖczyƒá (np. stworzyƒá nowƒÖ cechƒô "wiek klienta w momencie zakupu" z daty urodzenia i daty transakcji).

## Jak zrobiƒá dobry wykres?

üéØ Celowo≈õƒá: Skutecznie odpowiada na konkretne pytanie lub pokazuje kluczowy wzorzec. Dobry wykres ma jasny cel.

üëÅÔ∏è Czytelno≈õƒá: Jest ≈Çatwy do zrozumienia na pierwszy rzut oka, bez konieczno≈õci d≈Çugiego studiowania.

‚öñÔ∏è Rzetelno≈õƒá: Uczciwie i dok≈Çadnie reprezentuje dane, nie wprowadzajƒÖc w b≈ÇƒÖd (np. przez manipulacjƒô skalami osi).

‚ú® Prostota i Minimalizm: Skupia siƒô na danych, eliminujƒÖc zbƒôdne elementy graficzne ("≈õmieci wykresowe", ang. chartjunk), kt√≥re rozpraszajƒÖ uwagƒô.

üè∑Ô∏è Kontekst: Posiada wszystkie niezbƒôdne elementy: jasny tytu≈Ç, opisane osie (wraz z jednostkami) oraz czytelnƒÖ legendƒô (je≈õli jest wymagana).

üõ†Ô∏è Odpowiedni typ: Rodzaj wykresu (np. s≈Çupkowy, liniowy, punktowy) jest w≈Ça≈õciwie dobrany do rodzaju danych i historii, kt√≥rƒÖ ma opowiedzieƒá.

```python
# Plotting Iris petal length vs. petal width, colored by species
plt.figure(figsize=(8, 6))
sns.scatterplot(data=iris_df, x='petal length (cm)', y='petal width (cm)', hue='species', palette='viridis')
plt.title('Iris Petal Length vs. Petal Width by Species')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')
plt.legend(title='Species')
plt.show()
```

```python
# Plotting the distribution of the target variable (SalePrice)
plt.figure(figsize=(10, 6))
plt.hist(housing_data[htarget], bins=30)
plt.title('Rozk≈Çad zmiennej SalePrice')
plt.xlabel('SalePrice')
plt.ylabel('Liczebno≈õƒá')
plt.show()
```

```python
sns.histplot(banking_data, x='duration', hue=btarget, log_scale=True, bins=100, stat='probability', common_norm=False, alpha=0.5)
plt.xlabel('balance')
plt.ylabel('liczebno≈õƒá')
plt.title('Histogram zmiennej `duration` w zale≈ºno≈õci od pomy≈õlno≈õci rozmowy')
plt.tight_layout()
plt.show()
```

## Typy wykres√≥w



### Wykres punktowy (Scatter plot)

Wykres punktowy pokazuje zwiƒÖzek miƒôdzy dwiema zmiennymi ilo≈õciowymi, wy≈õwietlajƒÖc ka≈ºdy punkt danych jako kropkƒô.

```python
# Kod dla wykresu punktowego

plt.figure(figsize=(8, 6))
plt.scatter(x=iris_df.loc[iris_df[itarget] == 0, 'petal length (cm)'],
            y=iris_df.loc[iris_df[itarget] == 0, 'petal width (cm)'],
            label=iris.target_names[0])
plt.scatter(x=iris_df.loc[iris_df[itarget] == 1, 'petal length (cm)'],
            y=iris_df.loc[iris_df[itarget] == 1, 'petal width (cm)'],
            label=iris.target_names[1])
plt.scatter(x=iris_df.loc[iris_df[itarget] == 2, 'petal length (cm)'],
            y=iris_df.loc[iris_df[itarget] == 2, 'petal width (cm)'],
            label=iris.target_names[2])
plt.title('Iris Petal Length vs. Petal Width by Species')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')
plt.legend(title='Species')
plt.show()
```

### Wykres liniowy

Wykres liniowy s≈Çu≈ºy do przedstawienia zmian warto≈õci w czasie lub wzd≈Çu≈º innej ciƒÖg≈Çej zmiennej.

```python
# Kod dla wykresu liniowego

# Combine 'MoSold' and 'YrSold' to create a datetime column
housing_data['SaleDate'] = pd.to_datetime(housing_data['YrSold'].astype(str) + '-' + housing_data['MoSold'].astype(str).str.rjust(2, "0"))

# Display the first few rows with the new column
display(housing_data[['YrSold', 'MoSold', 'SaleDate']].head())
```

```python
sales = housing_data.groupby('SaleDate').count()['Id']

plt.figure(figsize=(12, 6)) # Added figure size for better readability
plt.plot(sales.index, sales, marker='o') # Corrected to use sales.index and sales
plt.title('Liczba sprzeda≈ºy w czasie') # Added a title
plt.xlabel('MiesiƒÖc sprzeda≈ºy') # Added x-axis label
plt.ylabel('Liczba sprzeda≈ºy') # Added y-axis label
plt.grid(True) # Added a grid for better readability
plt.show()
```

### Wykres s≈Çupkowy (Bar plot)

Wykres s≈Çupkowy por√≥wnuje warto≈õci dla r√≥≈ºnych kategorii, przedstawiajƒÖc je w postaci s≈Çupk√≥w o wysoko≈õciach proporcjonalnych do warto≈õci.

```python
# Kod dla wykresu s≈Çupkowego

nans = housing_data.groupby('MoSold').apply(lambda df: df['LotFrontage'].isna().mean(), include_groups=False)
plt.bar(np.arange(12) + 1, nans * 100)
plt.xticks(np.arange(12) + 1)
plt.title('Udzia≈Ç pustych warto≈õci zmiennej `LotFrontage`')
plt.xlabel('miesiƒÖc')
plt.ylabel('procent pustych warto≈õci')
plt.show()
```

### Histogram

Histogram przedstawia rozk≈Çad liczebno≈õci danych ciƒÖg≈Çych, dzielƒÖc zakres warto≈õci na przedzia≈Çy (kosze) i pokazujƒÖc liczbƒô obserwacji w ka≈ºdym przedziale.

```python
# Kod dla histogramu
plt.hist(housing_data.LotArea, bins=200)
plt.xlim([0, 25000])
plt.xlabel('LotArea')
plt.ylabel('Liczebno≈õƒá')
plt.title('Histogram zmiennej LotArea')
plt.show()
```

### Wykres pude≈Çkowy (Box plot)

Wykres pude≈Çkowy podsumowuje rozk≈Çad danych, pokazujƒÖc medianƒô, kwartyle oraz potencjalne warto≈õci odstajƒÖce.

```python
# Kod dla wykresu pude≈Çkowego

sns.boxplot(data=banking_data, y='balance', x=btarget)
plt.yscale('log')
plt.ylabel('balance')
plt.xlabel('Czy rozmowa zako≈Ñczy≈Ça siƒô sukcesem')
plt.title('Box plot zmiennej LotFrontage')
plt.show()
```

### Wykres skrzypcowy (Violin plot)

Wykres skrzypcowy ≈ÇƒÖczy cechy wykresu pude≈Çkowego z estymacjƒÖ gƒôsto≈õci rozk≈Çadu, pokazujƒÖc kszta≈Çt rozk≈Çadu danych.

```python
# Kod dla wykresu skrzypcowego

sns.violinplot(data=housing_data, x='SaleCondition', y=htarget)
plt.yscale('log')
plt.ylabel(htarget)
plt.xlabel('Sale Condition')
plt.title(f'Violin plot zmiennej {htarget}')
plt.show()
```

### Wizualizacja macierzy korelacji (Correlation Matrix Plot)

Macierz korelacji s≈Çu≈ºy do wizualizacji si≈Çy i kierunku zwiƒÖzku miƒôdzy parami zmiennych ilo≈õciowych. Czƒôsto przedstawiana jako mapa ciep≈Ça (heatmap).

```python
# Calculate the correlation matrix
correlation_matrix = housing_data.corr(numeric_only=True)

# Plot the heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, cmap='coolwarm', annot=False, center=0) # annot=True if you want to see the correlation values
plt.title('Macierz korelacji dla zbioru danych Housing')
plt.show()
```

### Inne typy wykres√≥w:

- Wykres radarowy (Radar chart)
  
  Wykres radarowy (lub pajƒôczy) s≈Çu≈ºy do por√≥wnywania wielu zmiennych ilo≈õciowych dla kilku grup lub punkt√≥w danych na wykresie dwuwymiarowym z osiami promieniowymi.

  https://plotly.com/python/radar-chart/

- Wykres tortowy (Pie chart)

  Wykres tortowy przedstawia proporcje poszczeg√≥lnych kategorii w stosunku do ca≈Ço≈õci w formie okrƒôgu podzielonego na sektory.

  https://matplotlib.org/stable/gallery/pie_and_polar_charts/pie_features.html

- Wykres konturowy (Contour plot)

  Wykres konturowy przedstawia funkcjƒô dw√≥ch zmiennych w postaci linii konturowych, ≈ÇƒÖczƒÖcych punkty o tej samej warto≈õci. Czƒôsto u≈ºywany do wizualizacji powierzchni 3D na p≈Çaszczy≈∫nie 2D.

  https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.contour.html

- Wykres kwantyl-kwantyl (Q-Q plot)

  Wykres kwantyl-kwantyl (Q-Q plot) s≈Çu≈ºy do por√≥wnania rozk≈Çadu pr√≥bki danych z teoretycznym rozk≈Çadem (np. normalnym), przedstawiajƒÖc kwantyle obu rozk≈Çad√≥w.

  https://matplotlib.org/mpl-probscale/tutorial/closer_look_at_viz.html
