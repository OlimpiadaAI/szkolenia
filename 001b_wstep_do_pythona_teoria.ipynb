## Obiekty, podstawowe typy i zmienne w Pythonie




Wszystko w Pythonie jest obiektem, a każdy obiekt w Pythonie ma typ. Niektóre z podstawowych typów to:

- `int` (liczba całkowita; liczba bez części dziesiętnej)
  - `10`
  - `-3`
- `float` (liczba zmiennoprzecinkowa; liczba z częścią dziesiętną)
  - `7.41`
  - `-0.006`
- `str` (ciąg znaków / łańcuch; sekwencja znaków zamknięta w pojedynczych, podwójnych lub potrójnych cudzysłowach)
  - `'to jest łańcuch używający pojedynczych cudzysłowów'`
  - `"to jest łańcuch używający podwójnych cudzysłowów"`
  - `'''to jest potrójnie cytowany łańcuch używający pojedynczych cudzysłowów'''`
  - `"""to jest potrójnie cytowany łańcuch używający podwójnych cudzysłowów"""`
- `bool` (wartość logiczna; wartość binarna, która jest albo prawdziwa, albo fałszywa)
  - `True`
  - `False`
- `NoneType` (specjalny typ reprezentujący brak wartości)
  - `None`

> W Pythonie zmienna to nazwa, którą określasz w swoim kodzie, mapująca się na konkretny obiekt, instancję obiektu lub wartość.

Definiując zmienne, możemy odnosić się do rzeczy po nazwach, które mają dla nas sens. Nazwy zmiennych mogą zawierać tylko litery, podkreślenia (`_`) lub cyfry (bez spacji, myślników ani innych znaków). Nazwy zmiennych muszą zaczynać się od litery lub podkreślenia.

## Podstawowe operatory

W Pythonie istnieją różne typy operatorów (specjalnych symboli), które działają na różnych wartościach. Niektóre z podstawowych operatorów to:

- operatory arytmetyczne
  - `+` (dodawanie)
  - `-` (odejmowanie)
  - `*` (mnożenie)
  - `/` (dzielenie)
  - `**` (potęgowanie)
- operatory przypisania
  - `=` (przypisz wartość)
  - `+=` (dodaj i przypisz ponownie; inkrementacja)
  - `-=` (odejmij i przypisz ponownie; dekrementacja)
  - `*=` (pomnóż i przypisz ponownie)
- operatory porównania (zwracają `True` lub `False`)
  - `==` (równy)
  - `!=` (nierówny)
  - `<` (mniejszy niż)
  - `<=` (mniejszy lub równy)
  - `>` (większy niż)
  - `>=` (większy lub równy)

Gdy w jednym wyrażeniu używanych jest wiele operatorów, kolejność operatorów określa, które części wyrażenia są oceniane w jakiej kolejności. Operatory o wyższym priorytecie są oceniane jako pierwsze (jak kolejność działań w matematyce). Operatory o tym samym priorytecie są oceniane od lewej do prawej.

- `()` nawiasy, do grupowania
- `**` potęgowanie
- `*`, `/` mnożenie i dzielenie
- `+`, `-` dodawanie i odejmowanie
- `==`, `!=`, `<`, `<=`, `>`, `>=` porównania

## Podstawowe kontenery

> Uwaga: obiekty stworzone z typów zmiennych można modyfikować po utworzeniu, a obiekty z typów niezmiennych nie.

Kontenery to obiekty, których można używać do grupowania innych obiektów. Podstawowe typy kontenerów to:

- `str` (łańcuch: niemutowalny; indeksowany liczbami całkowitymi; elementy są przechowywane w kolejności, w jakiej zostały dodane)
- `list` (lista: mutowalna; indeksowana liczbami całkowitymi; elementy są przechowywane w kolejności, w jakiej zostały dodane)
  - `[3, 5, 6, 3, 'pies', 'kot', False]`
- `tuple` (krotka: niemutowalna; indeksowana liczbami całkowitymi; elementy są przechowywane w kolejności, w jakiej zostały dodane)
  - `(3, 5, 6, 3, 'pies', 'kot', False)`
- `set` (zbiór: mutowalny; w ogóle nieindeksowany; elementy NIE są przechowywane w kolejności, w jakiej zostały dodane; może zawierać only obiekty niemutowalne; NIE zawiera zduplikowanych obiektów)
  - `{3, 5, 6, 3, 'pies', 'kot', False}`
- `dict` (słownik: mutowalny; pary klucz-wartość są indeksowane przez niemutowalne klucze; elementy NIE są przechowywane w kolejności, w jakiej zostały dodane)
  - `{'imię': 'Jadwiga', 'wiek': 23, 'ulub_jedzenie': ['pizza', 'owoce', 'ryba']}`

Podczas definiowania list, krotek lub zbiorów używaj przecinków (,), aby oddzielić poszczególne elementy. Podczas definiowania słowników używaj dwukropka (:), aby oddzielić klucze od wartości, i przecinków (,), aby oddzielić pary klucz-wartość.

Łańcuchy, listy i krotki to wszystko typy sekwencyjne, które mogą używać operatorów `+`, `*`, `+=` i `*=`.

## Atrybuty obiektów Pythona (metody i właściwości)

Różne typy obiektów w Pythonie mają różne **atrybuty**, do których można odwoływać się po nazwie (podobnie jak do zmiennej). Aby uzyskać dostęp do atrybutu obiektu, użyj kropki (`.`) po obiekcie, a następnie podaj atrybut (tj. `obiekt.atrybut`)

Kiedy atrybut obiektu jest wywoływalny, ten atrybut nazywa się **metodą**. Jest to to samo co funkcja, only ta funkcja jest związana z konkretnym obiektem.

Kiedy atrybut obiektu nie jest wywoływalny, ten atrybut nazywa się **właściwością**. Jest to po prostu fragment danych o obiekcie, który sam w sobie jest innym obiektem.

Wbudowanej funkcji `dir()` można użyć do zwrócenia listy atrybutów obiektu.



## Kilka metod na obiektach łańcuchowych (string)

- `.capitalize()` aby zwrócić wersję łańcucha z wielką pierwszą literą (only pierwszy znak wielką literą)
- `.upper()` aby zwrócić wersję łańcucha pisaną wielkimi literami (wszystkie znaki wielkimi literami)
- `.lower()` aby zwrócić wersję łańcucha pisaną małymi literami (wszystkie znaki małymi literami)
- `.count(podłańcuch)` aby zwrócić liczbę wystąpień podłańcucha w łańcuchu
- `.startswith(podłańcuch)` aby określić, czy łańcuch zaczyna się od podłańcucha
- `.endswith(podłańcuch)` aby określić, czy łańcuch kończy się podłańcuchem
- `.replace(stary, nowy)` aby zwrócić kopię łańcucha z wystąpieniami "starego" zastąpionymi przez "nowy"

```python
# Przypisz łańcuch do zmiennej
# Zostawiamy angielski łańcuch, aby przykłady .count() i .startswith() działały poprawnie
jakis_lancuch = 'tHis is a sTriNg'

# Zwróć wersję łańcucha z wielką pierwszą literą
jakis_lancuch.capitalize()

# Zwróć wersję łańcucha pisaną wielkimi literami
jakis_lancuch.upper()

# Zwróć wersję łańcucha pisaną małymi literami
jakis_lancuch.lower()

# Zauważ, że wywołane metody faktycznie nie zmodyfikowały łańcucha
jakis_lancuch

# Zlicz liczbę wystąpień podłańcucha w łańcuchu
jakis_lancuch.count('i')

# Zlicz liczbę wystąpień podłańcucha w łańcuchu po określonej pozycji
jakis_lancuch.count('i', 7)

# Zlicz liczbę wystąpień podłańcucha w łańcuchu
jakis_lancuch.count('is')

# Czy łańcuch zaczyna się od 'this'?
jakis_lancuch.startswith('this')

# Czy łańcuch pisany małymi literami zaczyna się od 'this'?
jakis_lancuch.lower().startswith('this')

# Czy łańcuch kończy się na 'Ng'?
jakis_lancuch.endswith('Ng')

# Zwróć wersję łańcucha z podłańcuchem zastąpionym czymś innym
jakis_lancuch.replace('is', 'XYZ')

# Zwróć wersję łańcucha z podłańcuchem zastąpionym czymś innym
jakis_lancuch.replace('i', '!')

# Zwróć wersję łańcucha z pierwszymi 2 wystąpieniami podłańcucha zastąpionymi czymś innym
jakis_lancuch.replace('i', '!', 2)
```

## Kilka metod na obiektach listowych (list)

- `.append(element)` aby dodać pojedynczy element do listy
- `.extend([element1, element2, ...])` aby dodać wiele elementów do listy
- `.remove(element)` aby usunąć pojedynczy element z listy
- `.pop()` aby usunąć i zwrócić element z końca listy
- `.pop(indeks)` aby usunąć i zwrócić element o podanym indeksie

```python
# Przypisz listę do zmiennej
jakas_lista = [1, 2, 3, 'witaj']

# Dodaj pojedynczy element do listy
jakas_lista.append(True)
print(jakas_lista)

# Dodaj wiele elementów do listy
jakas_lista.extend([4, 5, 6])
print(jakas_lista)

# Usuń pojedynczy element z listy
jakas_lista.remove('witaj')
print(jakas_lista)

# Usuń i zwróć element z końca listy
ostatni_element = jakas_lista.pop()
print(ostatni_element)
print(jakas_lista)

# Usuń i zwróć element o podanym indeksie
pierwszy_element = jakas_lista.pop(0)
print(pierwszy_element)
print(jakas_lista)
```

## Kilka metod na obiektach zbiorów (set)

- `.add(element)` aby dodać pojedynczy element do zbioru
- `.update([element1, element2, ...])` aby dodać wiele elementów do zbioru
- `.update(zbior2, zbior3, ...)` aby dodać elementy ze wszystkich podanych zbiorów do zbioru
- `.remove(element)` aby usunąć pojedynczy element ze zbioru
- `.pop()` aby usunąć i zwrócić losowy element ze zbioru
- `.difference(zbior2)` aby zwrócić elementy ze zbioru, których nie ma w innym zbiorze
- `.intersection(zbior2)` aby zwrócić elementy znajdujące się w obu zbiorach
- `.union(zbior2)` aby zwrócić elementy znajdujące się w którymkolwiek ze zbiorów
- `.symmetric_difference(zbior2)` aby zwrócić elementy, które są only w jednym zbiorze (nie w obu)
- `.issuperset(zbior2)` czy zbiór zawiera wszystko, co jest w innym zbiorze?
- `.issubset(zbior2)` czy zbiór jest zawarty w innym zbiorze?

```python
# Przypisz dwa zbiory do zmiennych
zbior_a = {1, 2, 3, 4, 5}
zbior_b = {4, 5, 6, 7, 8}

# Dodaj pojedynczy element do zbioru
zbior_a.add(0)
print(zbior_a)

# Dodaj wiele elementów do zbioru
zbior_a.update([1, 9, 10]) # Uwaga: 1 już jest w zbiorze, więc bez zmian
print(zbior_a)

# Usuń pojedynczy element ze zbioru
zbior_b.remove(8)
print(zbior_b)

# Usuń i zwróć losowy element ze zbioru
losowy_element = zbior_a.pop()
print(losowy_element)
print(zbior_a)

# Zwróć elementy ze zbioru_a, których nie ma w zbiorze_b
print(zbior_a.difference(zbior_b))

# Zwróć elementy znajdujące się w obu zbiorach
print(zbior_a.intersection(zbior_b))

# Zwróć elementy znajdujące się w którymkolwiek ze zbiorów
print(zbior_a.union(zbior_b))

# Zwróć elementy, które są only w jednym zbiorze (nie w obu)
print(zbior_a.symmetric_difference(zbior_b))

# Czy zbior_a zawiera wszystko, co jest w {1, 9}?
print(zbior_a.issuperset({1, 9}))

# Czy {4, 5} jest zawarte w zbiorze_b?
print({4, 5}.issubset(zbior_b))
```

## Kilka metod na obiektach słownikowych (dict)

- `.update([(klucz1, wart1), (klucz2, wart2), ...])` aby dodać wiele par klucz-wartość do słownika
- `.update(slownik2)` aby dodać wszystkie klucze i wartości z innego słownika do słownika
- `.pop(klucz)` aby usunąć klucz i zwrócić jego wartość ze słownika (błąd, jeśli klucza nie znaleziono)
- `.pop(klucz, wart_domyslna)` aby usunąć klucz i zwrócić jego wartość ze słownika (lub zwrócić wart_domyslna, jeśli klucza nie znaleziono)
- `.get(klucz)` aby zwrócić wartość dla określonego klucza w słowniku (lub None, jeśli klucza nie znaleziono)
- `.get(klucz, wart_domyslna)` aby zwrócić wartość dla określonego klucza w słowniku (lub wart_domyslna, jeśli klucza nie znaleziono)
- `.keys()` aby zwrócić listę kluczy w słowniku
- `.values()` aby zwrócić listę wartości w słowniku
- `.items()` aby zwrócić listę par klucz-wartość (krotek) w słowniku

```python
# Przypisz słownik do zmiennej
slownik_a = {'a': 1, 'b': 2, 'c': 3}

# Dodaj wiele par klucz-wartość do słownika
slownik_a.update([('d', 4), ('e', 5)])
print(slownik_a)

# Dodaj wszystkie klucze i wartości z innego słownika do słownika
slownik_a.update({'a': 100, 'f': 6}) # Uwaga: wartość 'a' zostaje nadpisana
print(slownik_a)

# Usuń klucz 'b' i zwróć jego wartość
wartosc = slownik_a.pop('b')
print(wartosc)
print(slownik_a)

# Usuń klucz 'z' i zwróć 'Nie dotyczy', jeśli nie znaleziono
wartosc = slownik_a.pop('z', 'Nie dotyczy')
print(wartosc)
print(slownik_a)

# Zwróć wartość dla klucza 'c'
print(slownik_a.get('c'))

# Zwróć wartość dla klucza 'z' (lub 'Nie znaleziono', jeśli nie znaleziono)
print(slownik_a.get('z', 'Nie znaleziono'))

# Zwróć listę kluczy
print(slownik_a.keys())

# Zwróć listę wartości
print(slownik_a.values())

# Zwróć listę par klucz-wartość
print(slownik_a.items())
```

## Argumenty pozycyjne i nazwane (keyword) w obiektach wywoływalnych

Funkcję/metodę można wywołać na wiele różnych sposobów:

- `func()`: Wywołaj `func` bez argumentów
- `func(arg)`: Wywołaj `func` z jednym argumentem pozycyjnym
- `func(arg1, arg2)`: Wywołaj `func` z dwoma argumentami pozycyjnymi
- `func(arg1, arg2, ..., argn)`: Wywołaj `func` z wieloma argumentami pozycyjnymi
- `func(nazwa_arg=wartosc)`: Wywołaj `func` z jednym argumentem nazwanym
- `func(nazwa_arg1=wartosc1, nazwa_arg2=wartosc2)`: Wywołaj `func` z dwoma argumentami nazwanymi
- `func(nazwa_arg1=wartosc1, nazwa_arg2=wartosc2, ..., nazwa_argn=wartosc_n)`: Wywołaj `func` z wieloma argumentami nazwanymi
- `func(arg1, arg2, nazwa_arg1=wartosc1, nazwa_arg2=wartosc2)`: Wywołaj `func` z argumentami pozycyjnymi i nazwanymi
- `obiekt.metoda()`: To samo co dla `func`.. i każdego innego przykładu `func`

Używając **argumentów pozycyjnych**, musisz je podać w kolejności, w jakiej funkcja je zdefiniowała (w **sygnaturze** funkcji).

Używając **argumentów nazwanych**, możesz podać argumenty, które chcesz, w dowolnej kolejności, o ile określisz nazwę każdego argumentu.

Używając argumentów pozycyjnych i nazwanych, argumenty pozycyjne muszą być pierwsze.

```python
# Zdefiniuj funkcję, która przyjmuje 3 argumenty
def wypisz_argumenty(arg1, arg2, arg3='wartość_domyślna'):
    print('arg1 to:', arg1)
    print('arg2 to:', arg2)
    print('arg3 to:', arg3)

# Wywołanie z argumentami pozycyjnymi
print('Wywołanie z argumentami pozycyjnymi:')
wypisz_argumenty('a', 'b', 'c')

# Wywołanie z argumentami nazwanymi (kolejność nie ma znaczenia)
print('\nWywołanie z argumentami nazwanymi:')
wypisz_argumenty(arg3='C', arg1='A', arg2='B')

# Wywołanie z argumentami pozycyjnymi i nazwanymi
print('\nWywołanie z mieszanymi argumentami:')
wypisz_argumenty('pierwszy', arg3='trzeci', arg2='drugi')

# Wywołanie używające wartości domyślnej dla arg3
print('\nWywołanie używające wartości domyślnej:')
wypisz_argumenty('jabłko', 'banan')

# To spowoduje błąd: argument pozycyjny po argumencie nazwanym
# wypisz_argumenty(arg1='x', 'y', 'z')
```

## Formatowanie łańcuchów i używanie placeholderów

```python
# "Stary" styl używający operatora %
imie = 'Alicja'
wiek = 30
print('Witaj, %s. Masz %d lat.' % (imie, wiek))

# "Nowy" styl używający metody .format()
print('Witaj, {}. Masz {} lat.'.format(imie, wiek))

# Metoda .format() z indeksami pozycyjnymi
print('Witaj, {1}. Masz {0} lat. Tak, {1}!'.format(wiek, imie))

# Metoda .format() z nazwami
print('Witaj, {n}. Masz {a} lat.'.format(n=imie, a=wiek))

# "Najnowszy" styl: f-stringi (Python 3.6+)
print(f'Witaj, {imie}. Masz {wiek} lat.')

# f-stringi mogą zawierać wyrażenia
print(f'Za 5 lat {imie} będzie mieć {wiek + 5} lat.')
print(f'Twoje imię wielkimi literami to {imie.upper()}.')
```

## Pętle "for" w Pythonie

Łatwo jest **iterować** po kolekcji elementów używając **pętli for**. Łańcuchy, listy, krotki, zbiory i słowniki, które zdefiniowaliśmy, są wszystko **iterowalnymi** kontenerami.

Pętla for przejdzie przez określony kontener, element po elemencie, i udostępni tymczasową zmienną dla bieżącego elementu. Możesz używać tej tymczasowej zmiennej jak normalnej zmiennej.

```python
# Pętla po liście
moja_lista = [1, 'a', True]
for element in moja_lista:
    print(f'Element to: {element}')

# Pętla po łańcuchu
for znak in 'witaj':
    print(znak.upper())

# Pętla po zakresie liczb
# range(5) generuje liczby 0, 1, 2, 3, 4
for i in range(5):
    print(f'Liczba {i}, Kwadrat to {i*i}')

# Pętla po słowniku (domyślnie iteruje po kluczach)
moj_slownik = {'owoc': 'jabłko', 'kolor': 'czerwony'}
print('\nPętla po kluczach słownika:')
for kl in moj_slownik:
    print(f'Klucz to "{kl}", a wartość to "{moj_slownik[kl]}"')

# Pętla po elementach słownika (parach klucz-wartość)
print('\nPętla po elementach słownika (klucz-wartość):')
for klucz, wartosc in moj_slownik.items():
    print(f'Klucz to "{klucz}", a wartość to "{wartosc}"')
```

## Instrukcje "if" i pętle "while" w Pythonie

Wyrażenia warunkowe mogą być używane z tymi dwiema **instrukcjami warunkowymi**.

**Instrukcja if** pozwala przetestować warunek i wykonać pewne działania, jeśli warunek jest prawdziwy (`True`). Możesz również dodać klauzule `elif` i/lub `else` do instrukcji if, aby podjąć alternatywne działania, jeśli warunek jest fałszywy (`False`).

**Pętla while** będzie się wykonywać dopóki jej wyrażenie warunkowe nie będzie fałszywe (`False`).

> Uwaga: Możliwe jest "zapętlenie w nieskończoność" podczas używania pętli while z wyrażeniem warunkowym, które nigdy nie staje się fałszywe.
>
> Uwaga: Ponieważ **pętla for** iteruje po kontenerze elementów, dopóki nie ma ich więcej, nie ma potrzeby określania warunku "zatrzymaj pętlę".

```python
# Przykład instrukcji if
x = 10

if x < 5:
    print('x jest małe')
elif x == 10:
    print('x to 10')
else:
    print('x jest duże')

# Warunki nie muszą być porównaniami
# Pusty kontener (lista, łańcuch itp.) jest traktowany jako "Fałsz"
moja_lista = []
if moja_lista:
    print('Lista nie jest pusta')
else:
    print('Lista jest pusta')

# Przykład pętli while
licznik = 0
while licznik < 5:
    print(f'Licznik to: {licznik}')
    licznik += 1  # Inkrementuj licznik
print('Pętla while zakończona.')

# Pętlę while można zatrzymać za pomocą 'break'
licznik = 10
while True: # To pętla w nieskończoność
    print(f'Odliczanie: {licznik}')
    licznik -= 1
    if licznik < 0:
        print('Start!')
        break # Wyjdź z pętli
```

## Wyrażenia listowe, zbiorowe i słownikowe (comprehensions)

```python
# "Długi" sposób tworzenia listy kwadratów
lista_kwadratow = []
for i in range(10):
    lista_kwadratow.append(i * i)
print(lista_kwadratow)

# "Skrócony" sposób (comprehension) tworzenia listy kwadratów
lista_kwadratow_skr = [i * i for i in range(10)]
print(lista_kwadratow_skr)

# Wyrażenie listowe z warunkiem
# Pobierz liczby parzyste od 0 do 19
parzyste = [x for x in range(20) if x % 2 == 0]
print(parzyste)

# Wyrażenie zbiorowe (automatycznie obsługuje duplikaty)
moje_liczby = [1, 2, 2, 3, 4, 5, 5, 5, 6]
zbior_kwadratow = {x * x for x in moje_liczby}
print(zbior_kwadratow)

# Wyrażenie słownikowe
# Stwórz słownik mapujący liczby na ich kwadraty
slownik_kwadratow = {x: x * x for x in range(5)}
print(slownik_kwadratow)

# Stwórz słownik z dwóch list
klucze = ['a', 'b', 'c']
wartosci = [1, 2, 3]
moj_slownik_skr = {k: v for k, v in zip(klucze, wartosci)}
print(moj_slownik_skr)
```

## Tworzenie obiektów z argumentów lub innych obiektów

Podstawowe typy i kontenery, których dotychczas używaliśmy, wszystkie dostarczają **konstruktorów typów**:

- `int()`
- `float()`
- `str()`
- `list()`
- `tuple()`
- `set()`
- `dict()`

Do tej pory definiowaliśmy obiekty tych wbudowanych typów, używając pewnych skrótów składniowych, ponieważ są one tak powszechne.

Czasami będziesz mieć obiekt jednego typu, który musisz przekonwertować na inny typ. Użyj **konstruktora typu** dla typu obiektu, który chcesz uzyskać, i przekaż obiekt, który aktualnie masz.

```python
# Konwertuj float na int (obcina część dziesiętną)
print(int(7.89))

# Konwertuj int na float
print(float(10))

# Konwertuj int na string
liczba_str = str(123)
print(liczba_str)
print(type(liczba_str))

# Konwertuj string na int
liczba_int = int('456')
print(liczba_int)
print(type(liczba_int))

# To spowoduje błąd ValueError
# int('witaj')

# Konwertuj listę na krotkę
moja_lista_dla_krotki = [1, 'a', True]
moja_krotka = tuple(moja_lista_dla_krotki)
print(moja_krotka)

# Konwertuj krotkę na listę
moja_lista = list(moja_krotka)
print(moja_lista)

# Konwertuj listę na zbiór (usuwa duplikaty)
lista_z_duplikatami = [1, 1, 2, 2, 3, 'a', 'a']
moj_zbior = set(lista_z_duplikatami)
print(moj_zbior)

# Konwertuj listę par klucz-wartość (krotek) na słownik
pary = [('imię', 'Bogdan'), ('wiek', 40)]
moj_slownik_z_listy = dict(pary)
print(moj_slownik_z_listy)
```
